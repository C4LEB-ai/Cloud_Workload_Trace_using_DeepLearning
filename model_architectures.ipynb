# GOOGLE MODEL ARCHITECTURES

The input shape (None, 5, 10) indicates that the model expects inputs of a certain shape during training:
- None: This represents the batch size, which is flexible. During model definition, TensorFlow doesn't enforce a specific batch size, so it's set to None, allowing the model to accept any batch size during training or inference.
- 5: This is the number of features per time step.
- 10: This is the sequence length, representing the number of time steps.
However, the input shape in your code should be (sequence_length, features), meaning (10, 5) if you follow the given variables. The discrepancy arises from how TensorFlow processes these inputs. The input shape (sequence_length, features) corresponds to (10, 5), but in the model plot, it's common to see the dimensions swapped due to how data is typically represented (with the batch size as the first dimension).

# ANN

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import plot_model

# Define input_shape
input_shape = (6,)  # Replace with the actual input shape from your first model

# Create the model
model = Sequential([
    Dense(64, activation='relu', input_shape=input_shape),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])

# Compile the model with the same settings
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])

# Plot the model
plot_model(model, to_file='dense_model.png', show_shapes=True)
```

# CNN

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Reshape
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import plot_model

# Define input_shape based on your training data
input_shape = (6,)  # Replace with the actual input shape

# Create the CNN model
cnn_model = Sequential([
    Reshape(target_shape=(6, 1), input_shape=input_shape),
    Conv1D(64, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model with a custom learning rate
custom_learning_rate = 0.001
optimizer = tf.keras.optimizers.Adam(learning_rate=custom_learning_rate)
cnn_model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])

# Plot the model
plot_model(cnn_model, to_file='cnn_model.png', show_shapes=True)
```

# ANN

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import plot_model

# Define the input shape based on your model
input_shape = (8,)  # Replace with the actual input shape

# Create the ANN model
model_ann = Sequential([
    Dense(128, activation='relu', input_shape=input_shape),
    Dropout(0.5),  # Dropout layer with a dropout rate of 0.5
    Dense(64, activation='relu'),
    Dense(1)
])

# Compile the model with a custom learning rate
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)
model_ann.compile(optimizer=optimizer, loss='mean_squared_error')

# Plot the model
plot_model(model_ann, to_file='ann_model.png', show_shapes=True)
```

# RNN

```python
import tensorflow as tf
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import plot_model

# Define the input shape based on your RNN model
input_shape = (8,1)  # Replace with the actual input shape

# Create the RNN model
model_rnn = Sequential([
    LSTM(64, input_shape=input_shape, activation='tanh', return_sequences=True),
    LSTM(64, activation='tanh'),
    Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    Dense(1)  # Output layer for regression
])

# Compile the model
model_rnn.compile(optimizer='adam', loss='mean_squared_error')

# Plot the model
plot_model(model_rnn, to_file='rnn_model.png', show_shapes=True)
```

# CNN

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.utils import plot_model

# Define input shape based on reshaped data
input_shape = (8, 1)  # Replace with the actual input shape

# Create the CNN model
model_cnn = models.Sequential([
    layers.Conv1D(32, kernel_size=3, activation='relu', input_shape=input_shape),
    layers.MaxPooling1D(pool_size=2),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
])

# Compile the model
model_cnn.compile(optimizer='adam', loss='mean_squared_error')

# Plot the model
plot_model(model_cnn, to_file='cnn_model.png', show_shapes=True)
```
